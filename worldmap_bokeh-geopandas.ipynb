```python
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure, show
from shapely.geometry import Polygon
import geopandas as gp
# this part is needed in case you run into a runtime error
#import bokeh.plotting as bop
#bop.output_notebook()

output_file("WorldMap.html")

world = gp.read_file(gp.datasets.get_path('naturalearth_lowres'))
names = [country for country in world.name]

countries = []
[countries.append(country) if type(item) == Polygon else [countries.append(country) for i in list(item)] for
 item, country in zip(world.geometry, names)]

polygons = []
[polygons.append(item) if type(item) == Polygon else [polygons.append(i) for i in list(item)] for item in
 world.geometry]

xs, ys = [], []
for i,polygon in enumerate(polygons):
    try:
        xs.append(list(polygon.boundary.coords.xy[0]))
        ys.append(list(polygon.boundary.coords.xy[1]))
    except NotImplementedError:
        print('CATCH')
        num = i
        continue

countries.pop(num)
source = ColumnDataSource(dict(xs=xs, ys=ys, countries=countries))

p = figure(title='World', tools='pan, wheel_zoom, box_zoom, reset, hover, save', tooltips=[('Countries', '@countries')],
           x_range=(-180, 180), y_range=(-90, 90), x_axis_location=None, y_axis_location=None,plot_width=1000,plot_height=500)

p.patches('xs', 'ys', fill_alpha=0.4, fill_color='grey', line_color='black', line_width=0.5, source=source)

# this part is needed as I am running into a runtime error.
# found this solution here https://github.com/bokeh/bokeh/issues/8579
# uncomment the following
#try:
    #bop.reset_output()
    #bop.output_notebook()
    #bop.show(p)  # angrily yells at me about single ownership
#except:
    #bop.output_notebook()
    #bop.show(p)  # ...aaaaaaand we're back to happy inline plotting.

# comment this out if you are using the block from 43-49   
show(p)

```
